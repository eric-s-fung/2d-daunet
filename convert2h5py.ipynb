```python
import os
import glob
import nibabel as nib
import h5py 
```

```python
img_train = glob.glob("C:/users/fungs/bmen501/data/Train/Image/*.nii.gz")
mask_train = glob.glob("C:/users/fungs/bmen501/data/Train/Mask/*.nii.gz")
robex_train = glob.glob("C:/users/fungs/bmen501/data/Train/Mask/*mask.nii")
img_val = glob.glob("C:/users/fungs/bmen501/data/Val/Image/*.nii.gz")
mask_val = glob.glob("C:/users/fungs/bmen501/data/Val/Mask/*.nii.gz")
robex_val = glob.glob("C:/users/fungs/bmen501/data/Val/Mask/*mask.nii")
nifti_files = img_train + mask_train + img_val + mask_val
print(robex_val)
```

```python
for ii in nifti_files:
    out_file = ii.split("/")
    out_file[-2] = out_file[-2] + "-hdf5"
    out_file[-1] = out_file[-1].split(".nii.gz")[0] + ".h5"
    out_file[0] = out_file[0] + "\\"
    out_file = os.path.join(*out_file)
    print(out_file)
    img = nib.load(ii).get_fdata()
    print(img.shape)
    img = (img - img.min())/(img.max() - img.min()+1e-8)
    if out_file.split("\\")[-1].startswith("C"):
        hf = h5py.File(out_file, 'w')
        hf.create_dataset('data', data = img.transpose(2,0,1))
        hf.close()
    else:
        hf = h5py.File(out_file, 'w')
        hf.create_dataset('data', data = img.transpose(1,0,2))
        hf.close()
        
```
